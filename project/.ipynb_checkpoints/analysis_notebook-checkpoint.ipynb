```python
# Aggregate respiratory data by state and year
state_respiratory = respiratory_df.groupby(['state_code', 'year'])['respiratory_value'].median().reset_index()

# Aggregate air quality data by state and year
state_air_quality = air_quality_df.groupby(['state_code', 'year'])['sample_measurement'].median().reset_index()

# Merge the two datasets
state_merged = pd.merge(state_respiratory, state_air_quality, on=['state_code', 'year'], how='inner')

print("\nState-Level Merged Data:")
print(state_merged.head())
```

```python
# Calculate correlation for state-level data
state_pearson_corr, _ = pearsonr(state_merged['sample_measurement'], state_merged['respiratory_value'])
print(f"State-Level Pearson Correlation: {state_pearson_corr:.2f}")

# Regression for state-level data
X_state = state_merged[['sample_measurement']]
y_state = state_merged['respiratory_value']

# Fit regression model
model_state = LinearRegression()
model_state.fit(X_state, y_state)

# Predict and calculate R² score
y_state_pred = model_state.predict(X_state)
state_r2 = r2_score(y_state, y_state_pred)
print(f"State-Level R² Score: {state_r2:.2f}")
print(f"State-Level Regression Coefficient: {model_state.coef_[0]:.2f}")
print(f"State-Level Regression Intercept: {model_state.intercept_:.2f}")
```

```python
# Plot state-level trends
plt.figure(figsize=(12, 6))
plt.scatter(state_merged['sample_measurement'], state_merged['respiratory_value'], color='blue', alpha=0.6)
plt.plot(state_merged['sample_measurement'], y_state_pred, color='red', label='Regression Line')
plt.title('State-Level PM2.5 vs Respiratory Diseases')
plt.xlabel('PM2.5 Median (Micrograms/m^3)')
plt.ylabel('Respiratory Median Value')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Calculate year-over-year changes
merged_df['pm25_change'] = merged_df['pm25_avg'].diff()
merged_df['resp_change'] = merged_df['respiratory_value'].diff()

# Drop NaN values (from diff)
merged_changes = merged_df.dropna(subset=['pm25_change', 'resp_change'])

# Calculate correlation for year-over-year changes
change_corr, _ = pearsonr(merged_changes['pm25_change'], merged_changes['resp_change'])
print(f"Yearly Change Correlation between PM2.5 and Respiratory Value: {change_corr:.2f}")

# Scatter plot for year-over-year changes
plt.figure(figsize=(10, 6))
plt.scatter(merged_changes['pm25_change'], merged_changes['resp_change'], color='green')
plt.title('Yearly Change in PM2.5 vs Respiratory Diseases')
plt.xlabel('Yearly Change in PM2.5 (Micrograms/m^3)')
plt.ylabel('Yearly Change in Respiratory Value')
plt.grid(True)
plt.show()
```
